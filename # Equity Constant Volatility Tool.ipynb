## Constant Volatility calibration overview

### Constant Volatility Output models

The **Equity Constant Volatility** tool, as we will call it, is our first *multi-asset* tool. The current run (end-September 2017) produces a calibration for 188 equity assets as one single operation.  

The tool generates model calibrations per asset for:

+ *Assets.EQ.PEA*, colloquially, the **Factor Beta Calibration**
    + Factor Beta == Factor Loading - these are the same thing and are used interchangeably
+ *Assets.EQ.PEA.Sigma.Const*, aka the **Constant Volatility Calibration**
    + The additional outputs are carried by this model - i.e. calibration statistics go here.
    
We also need to produce a calibration for the *Assets.EQ.Factors.Sigma.Const* model - the **Factor Volatility Calibration**, that the current XLS tool generates. We also need this as an input to the tool, so no actual calibration is going on here. This is a perfect candidate for being added to the DB as a **persistent, static** model. I have done just that in the **Assumptions_IE_Defaults.xlsm** export tool, and this model will be avaliable as one of our tool model inputs. We should stop creating a copy of this model each calibration run, and instead use this persistent copy in BHC creation.

### Constant Vol ESG Validation

With a new set of calibrations, we need to run this through the ESG to produce some supporting validation results. Unlike the excel tool, which performs this validation as part of the tool run, our new tool run will **not** include a validation run or any validation results.  

We will need to create a separate **RW Equity Validation** tool, that will take the results from this tool and perform whatever ESG runs we need. A separate specification for this tool will be on its way soon...

> Note: Validation is pretty important - we will need to replicate this to consider the whole feature complete - just not as part of the ConstVol tool itself. 

### Model Inputs

For our calibration here, we need to enter:

+ **Asset Unconditional Volatility Targets** - the target *Total Volatility* for the asset
+ **Asset Correlation Targets** - a *matrix* of correlations between each asset pair
+ the aforementioned **Factor Volatility Calibration**, which will be used to determine the *Systematic Volatility* as the product of the factor vols and the **Factor Beta Calibration**
+ A **Settings Model**, which will provide 
    + all of the global options that we want to use in the calibration.
    + configuration options for *each asset* that we need to set, which will also serve as a list of all the assets we need to fit.

Now is a good time to give a very quick overview of how the calibration works:

+ Find the **Factor Beta Calibration** for each asset that minimises the error in the **Asset Correlation Targets** matrix
    + Sneak preview - there are two ways in which we do this
+ Calculate the *Systematic Volatility* that the new **Factor Beta Calibration** implies.
+ Calculate the *Specific Volatility* that will work on top of *Systematic Volatility* to hit the target *Total Volatility*, from the **Asset Unconditional Volatility Targets**  - this result is the **Constant Volatility Calibration**

So far, so simple right? Time to make this more complicated.

## Constant Volatility Calibration - in detail

### Targets are generated per-asset and for all-asset

Correction - we intend our mercury version of **RWTS** to produce results *per-asset* where appropriate. This means that we produce a single model for each asset.  

Where not appropriate - i.e. correlations, where we are cataloguing the relationships between assets, we produce targets that are *all-asset* - i.e. we collect together results for all assets in a single model.  

Either approach to storing models is fine, but have their drawbacks. For *per-asset*, we can accept new assets easily, but the TMT configuration will be unwieldy (at least 188 input models for end-September). For *all-asset* the config will be more straightforward, but we will need to update the single model whenever we get new information (new assets) which could trigger downstream runs that we don't want.  

For the specifcation of the new tool here, I will assume that we will use an *all-asset* approach for all model inputs, and assume anything that we need as an input that is *per-asset* has been sorted elsewhere into a single *all-asset* model.  

I have created a tool [Assumptions_RWTS_Import_Export_ConstVolTargets.xlsm](U:\CS\2017 Projects\RWTS\Mock Import-Export\Assumptions_RWTS_Import_Export_ConstVolTargets.xlsm) to do this. When we complete on the **RWTS** project we should reconsider how we pass in models, and which tool is responsible for combining and sorting data in readyness for calibration. For now, the *all-asset* approach is closer to how the tool method will handle data, so this makes for the most straightforward convertion to get the tool working.  

### Not all quarterly calibrations are the same

What I have described so far is the *Full Calibration* where we calibrate both the **Factor Beta Calibration** to new targets, and then calculate the **Constant Volatility Calibration** that stems from the factor fit.  

The targets that we hit in this calibration are generated by the **Real World Target Setting** process - you may have heard of it, and its rather odd cycles. As a result, the targets that we want to hit are:

+ Updated every six months
    + End-March and End-September
+ Delayed by three months
    + To give clients notice of updates
    + So we actually use End-June and End-December data
+ Subject to change, when we actually complete on converting **RWTS** to a mercury process
    + We may go monthly, go annual, who knows?
    
So, we have in addition to the *Full Calibration* of updates, we also need a mode for an *Update Calibration*.  

To do this, we will need to ensure that the tool can take an existing **Factor Beta Calibration** as an input, and "fix" the calibration for all assets that were calibrated previously - i.e. the model output should be accepted as a model input.  

In practice we will represent this in this specification as an output *all-asset* version of the **Factor Beta Calibration**, and a corresponding, optional *all-asset* **Factor Beta Calibration** as an input. When this input model is used, the tool will need to be smart enough to adjust settings such that the existing **Factor Beta Calibration** is retained for each asset, and only *new* assets not present in the past result are calibrated.  

Note that the **Constant Volatility Calibration** is a simple function of the **Factor Beta Calibration** and the **Factor Volatility Calibration**, so we can (re)calculate these values each time.  

> The delay and irregular updates will make this even trickier. To handle this, we may require a *sorter tool* to take the non-persistent factor output model > from this tool, and create a *persistent* copy of the model with a different criteria. We can then run the sorter infrequently to create the effect of irregular updates that we need. A final answer on our future schedule for RWTS updates would be handy before we decide to build this part.

This will mean that we are assuming that we always run a *Full Calibration*, but we should also test that an *Update Calibration* will work in our first release.  

### There are two methods that we need to run in sequence

To calculate the **Factor Beta Calibration** with the **MoodysAnalytics.Calibration.Tool.RealWorldEquity** tool, we in practice need to run two separate methods in sequence:  

+ First we, need to run the **PCA** (principal components analysis) on the first N assets added to the tool.
    + So, there is a strict order that we need to keep for the list of assets that we calibrate to.
    + We will pass the value for N in with the **Settings Model**.
+ Next, we perform a **numeric fit** -  i.e. we optimise - to find the fit for the remaining assets in sequence (for N+1 -> total #assets).
    + The **PCA** fit for the first N assets is an *input* to this method - it effects the objective function for each optimisation.
    + Again, order is important.
    
Digging into the code base for the Real World Equity tool, we may be shocked and amazed to discover that we need to implement two separate tools:

+ The **PCA** is the **FactorSensitivityCalculation** class in the *MoodysAnalytics.Calibration.Tool.RealWorldEquity.EquityFactorTool.InitialFactorReduction* namespace
+ The **numeric fit** is the **NumericalMinimizationCalculation** class in the *MoodysAnalytics.Calibration.Tool.RealWorldEquity.EquityFactorTool.NumericalMinimization* namespace

These are traditional Tools, in the sense that:

+ They have their own data contract for inputs and outputs.
+ They have a "Calculate" method that we call to calibrate the tool.

If we want to create a single **Equity Constant Volatility** tool we want to chain these two methods above to create a single tool that mercury can interact with. We can explore how we want to wrap these up and expose the methods independently later in this spec, when we start drawing TMT diagrams.  

### The outputs are incomplete

We will return output parameters from the tool methods which only partially covers what we currently get from the excel constant vol tool. These are avaliable as methods in the **MoodysAnalytics.Calibration.Tool.RealWorldEquity** tool code base, so we will need to pass our calibration result to these methods to complete the output set.  

#### Specific Vol output

Most glaringly the *specific vol* that constitutes the **Constant Volatility Calibration** is not part of the outputs, and does not have a method within the code base! We will need to jump through the following flaming hoops to get to the specific volatility number:

+ *Total Volatility* is the input from the **Asset Unconditional Volatility Targets**
    + which we can denote as $ \sigma_{Total} $
+ *Systematic Volatility* is a function of the **Factor Beta Calibration** we fit here, and the **Factor Volatility Calibration**.
    + We can find this function in the code base as the method **CalculateSystematicVol**, in the class *GeneralEquityCorrelationCalcs* in the namespace *MoodysAnalytics.Calibration.Tool.RealWorldEquity.EquityBondCorrelation*
    + we can denote this as $ \sigma_{Systematic} $
+ *Specific Volatility* we will need to derive ourselves:
    + $ \sigma_{Specific} = \sqrt{\sigma_{Total}^2 - \sigma_{Systematic}^2 }$
    + in english = the square root of the difference of the squares
+ There may be cases where we could go lower than a specified minimum specific volatility  $ \sigma_{Minimium} $
    + This is declared in settings, and is passed to the factor beta calibration
    + To account for this, we should use the following, final formula for *Specific Volatility*
        + $ \sigma_{Specific} = \sqrt{\max{\{\sigma_{Minimium}^2, \sigma_{Total}^2 - \sigma_{Systematic}^2 \}}}$ 
    + So in these cases, $ \sigma_{Specific} = \sigma_{Minimium} $
    
So straightforward enough, right?  

To summarise, we will create **Total Vol**, **Specific Vol** and **Systematic Vol** parameters that we can use in tool model output.  

#### Correlation Estimate output

We will also need to garnish the outputs with the *model correlation approximation*. This can be found in the code base - the one to use is the matrix result member **CorrelationMatrixApproximationWithDiagonalReplacement** of the *CorrelationMatrixApproximationCalcs* class, which lives in the *MoodysAnalytics.Calibration.Tool.RealWorldEquity.EquityFactorTool.NumericalMinimization* namespace.  

In addition, we will need to set this up so that we use a pair of asset names to index each value in the matrix that we push out as a model.  

We will need to pass in data from the output models of the method to return a single matrix (a *double[,]*) parameter, **Model.Output.CorrelationMatrix** that we can push out to the output model results.

#### PCA Cumulative result

From the PCA fit, we output results on the Eigenvalue decomposition per-eigenvalue. We always show this in documentation as a percentage of volatility described by the  eigenvalues *up to the current eigenvalue*, so it would be wise to calculate this within the outputs.  

We want to take the *cumulativeVarience*[sic] output from the PCA outputs, and calculate from this an array of the **percentage cumulative variance**.  

For the definition of this, I will steal from the excel UI code wholesale, and remove the parts we already have:

``` C#

            var N = PCA.NumAssets 
            var cumVar = new double[N]; // This is the "percentage cumulative variance" we want to return
            var nFactors = new double[N];
            var sumOfEigenValues = calculationResults.Outputs.OrderedEigenValues.Sum();

            var eignenValuesSoFar = 0.0;

            for (int i = 0; i < N; i++)
            {
                eignenValuesSoFar = eignenValuesSoFar + calculationResults.Outputs.OrderedEigenValues[i];
                cumVar[i] = eignenValuesSoFar/sumOfEigenValues;
                nFactors[i] = i + 1;                

            }         

```

We can then assume that we have a **percentage cumulative variance** output with a correctly calculated value, that can be appended to the original *orderedEigenValues* and *factorVols* results that we can add to the tool output - we will refer to this later as the **PCA Cumulative Result**.  

This will consist of a *Table* styled output with the headings **Index, EigenValue, PctCumVar & FactorVol** - so like the above attempt to write data, but with an additional **Index** (equal to "i" in the above) so we can preserve order.  

### Outputs are required per-asset and for all-asset

### PCA Calibration mapping

Quick recap: The **PCA Calibration** part of the tool is the **FactorSensitivityCalculation** class in the *MoodysAnalytics.Calibration.Tool.RealWorldEquity.EquityFactorTool.InitialFactorReduction* namespace.
